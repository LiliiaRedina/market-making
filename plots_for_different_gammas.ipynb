```python
import matplotlib.pyplot as plt
f = open("results.txt")
ss = f.read().split("#")
gs = [1, 0.7, 0.5, 0.2, 0.1, 0.05, 0.01]
coins = []
cashs = []
commons = []
for s in ss:
    if s == '':
        continue
    lines = s.split('\n')
    # print(len(lines))
    # assert len(lines) == 6
    gamma = lines[0]
    try:
      cashs.append(([float(x) for x in lines[1].split(" ") if x != ""], gamma))
      coins.append(([float(x) for x in lines[2].split(" ") if x != ""], gamma))
      commons.append(([float(x) for x in lines[3].split(" ") if x != ""], gamma))
    except (Exception):
      print(gamma)

for ys, label in cashs:
    plt.plot(ys, label=label)
plt.legend()
plt.title("cash")
plt.show()

for ys, label in coins:
    plt.plot(ys, label=label)
plt.legend()
plt.title("coin")
plt.show()

for ys, label in commons:
    plt.plot(ys, label=label)
plt.legend()
plt.title("common")
plt.show()

```
